```python
def floods_warning_function(yyyy_mm_dd):
   
    # collecting the data
    
    url = 'https://newsapi.org/v2/everything?'
    param = {
        'q': 'flood',
        'apiKey' : 'e685d6e1420f4882b86d029ed3c1a11d',
        'from': yyyy_mm_dd,
        #'to': '2019-04-15',
        'pageSize': 100,
        'language': 'en',
        'sortBy':'relevancy',
        'sources':'the-washington-post, the-new-york-times, reuters, \
        nbc-news, msnbc, google-news, fox-news, cnn, cbs-news, abc-news, associated-press'
    }
    articles = []
    for page in range(1,20):
        param['page'] = page
        floods = requests.get(url, params = param)
        floods_articles = floods.json()['articles']
        articles.extend(floods_articles)
        if len(floods_articles) ==0:
            break    
    print(f'There are {len(articles)} articles containing the word "flood" since {yyyy_mm_dd}.')
    print(' ')


    # putting it together into a dataframe
    
    df_new = pd.DataFrame(articles)
    df_new['source_id'] = df_new['source'].map(lambda x: x['id'])
    df_new['source_name'] = df_new['source'].map(lambda x: x['name'])
    df_new.drop(columns = ['source','urlToImage','source_id','author'], inplace=True)
    
    
    # Anna's model:
    # Importing and transforming train data:
    df =pd.read_csv('./floods_relevance.csv')
    df3 = df_new
    df.rename(columns={'flood_relevance':'positive'}, inplace=True)
    stop = list(text.ENGLISH_STOP_WORDS)
    #adding stop words, including country and state names
    stop.extend(['playing','said', 'Image', 'copyright', 'Media', 'caption'])
    stop.extend(us_states_list)
    stop.extend(country_list)
    tvec=TfidfVectorizer(stop_words=stop, min_df=5, max_df=1.0, ngram_range=(1,2), max_features=5000)
    svd=TruncatedSVD(random_state=42, n_components=100)
    df['all_text'] = df['title']+' '+ df['description']+ ' ' + df['content']
    df3['all_text'] = df3['title']+' ' +df3['description']+ ' ' +df3['content']
    def preprocess (dataframe):
        dataframe.drop_duplicates(inplace=True)
        dataframe.dropna(inplace=True)
        dataframe.iloc[:,0].replace('\d+', ' ', regex=True, inplace=True)
        dataframe.reset_index(inplace=True)
        return dataframe
    df= preprocess(df)
    df3 = preprocess(df3)


    # Initialize and Fit Training Data:
    cont_mat=tvec.fit_transform(df['all_text'])
    cont_df=pd.SparseDataFrame(cont_mat, columns=tvec.get_feature_names())
    cont_df.fillna(0, inplace=True)
    svd_matrix=svd.fit_transform(cont_df)
    comp_names=['component_'+str(i+1) for i in range(100)]
    svd_df=pd.DataFrame(svd_matrix, columns=comp_names)

    
    # Transform the Unseen Test Data:
    cont_mat3=tvec.transform(df3['all_text'])
    cont_df3=pd.SparseDataFrame(cont_mat3, columns=tvec.get_feature_names())
    cont_df3.fillna(0, inplace=True)
    svd_test = svd.transform(cont_df3)
    comp_names=['component_'+str(i+1) for i in range(100)]
    svd_df3=pd.DataFrame(svd_test, columns=comp_names)

    # Fitting RandomForest
    X=df['all_text']
    y=df['positive']
    X_train, X_test, y_train, y_test = train_test_split(svd_df, y, stratify=y, random_state = 42)
    rfc1=RandomForestClassifier(random_state=42, max_depth=4, min_samples_split=3, n_estimators=40)
    rfc1.fit(X_train, y_train)
    
    
    # Predictions on unseen data
    preds=rfc1.predict(svd_df)
    preds3=rfc1.predict(svd_df3)
    df3['preds3']=preds3
    
    # Displaying relevant information 
    print(f"{df3['preds3'].sum()} are flagged as relevant to a disaster.")
    print(' ')
    
    # prints title name, url, and relevant country or state
    title_list = []
    for i in range(len(df3['preds3'])):
        if df3.loc[i, 'preds3'] ==1:
            title_list.append(df3.loc[i, 'title'])
            print(f'Article {i} title:', df3.loc[i, 'title'])
            print('url', df3.loc[i,'url'])
            for country in country_list:
                country_hold = []
                if country in df3.loc[i, 'all_text']:
                    country_hold.append(country)
                if len(country_hold)>0:
                    print('Relevant Country: ', country_hold)
            for state in us_states_list:
                state_hold = []
                if state in df3.loc[i, 'all_text']:
                    state_hold.append(state)
                if len(state_hold)>0:
                    print('Relevant State(s): ', state_hold)
            print(' ')
    
    # data frame with repeated titles and their frequency
    title_list = []
    for i in range(len(df3['preds3'])):
        if df3.loc[i, 'preds3'] ==1:
            title_list.append(df3.loc[i, 'title'])
    pd.set_option('display.max_colwidth', -1)
    output_df = pd.DataFrame()
    output_df['Article Title'] = list(set(title_list))

    article_relevance = []
    for title in list(set(title_list)):
        article_relevance.append((title_list).count(title))
            
    output_df['Relevance (number of articles)'] =  article_relevance
    return output_df.sort_values(by = ['Relevance (number of articles)'], ascending = False)
    
            
```
